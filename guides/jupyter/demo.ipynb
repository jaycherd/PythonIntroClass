# Welcome to Python Basics Demo

This Jupyter Notebook will introduce you to the basics of Python, including simple data types and operations.

```python
# Print statements are used to output data to the standard output device (screen).
print("Hello, Python!")

# You can comment your code like this. Comments are not executed.
```

```python
# Variables are containers for storing data values.
a = 5        # integer
b = 3.14     # float
c = "Python" # string

# Let's print the variables
print(a, b, c)
```

```python
# Python supports the basic arithmetic operations
sum = a + 1   # addition
diff = a - 1  # subtraction
prod = a * 2  # multiplication
quot = a / 2  # division

# Printing the results
print("Sum:", sum)
print("Difference:", diff)
print("Product:", prod)
print("Quotient:", quot)
```

```python
# Lists are used to store multiple items in a single variable.
my_list = [1, 2, 3, 4, 5]

# Looping through a list
for number in my_list:
    print(number)

print(f"sum: {sum}")
```

```python
# A function is a block of code which only runs when it is called.
def greet(name):
    return "Hello " + name + "!"

# Calling the function
greeting = greet("Alice")
print(greeting)
```

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Data (hypothetical)
signal_present = np.random.normal(loc=0.75, scale=0.1, size=100)  # Hit rates
signal_absent = np.random.normal(loc=0.3, scale=0.1, size=100)    # False alarm rates

# Create histograms of the two distributions
plt.figure(figsize=(12, 6))
plt.hist(signal_present, bins=20, alpha=0.7, label='Signal Present (Hits)')
plt.hist(signal_absent, bins=20, alpha=0.7, label='Signal Absent (False Alarms)')

# Customizing the plot
plt.title('Signal Detection Theory: Hits vs. False Alarms')
plt.xlabel('Rate')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```
